# Application of KMeans to quantify images, colours etc etc 

(HEIGHT, WIDTH, COLOR CHANNELS)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import matplotlib.image as mpimg 
#.jpg .png ---> convert ----> numpy array 
```

```python
image_as_array = mpimg.imread('DATA/palm_trees.jpg')
```

```python
image_as_array.shape
```

```python
image_as_array
```

```python
plt.imshow(image_as_array)
```

```python
#tuple unpacking 
(h,w,c) = image_as_array.shape
```

```python
h
```

```python
image_as_array2d = image_as_array.reshape(h*w,c)
```

```python
image_as_array2d
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=6) #quantising to approx 6 colors
```

```python
labels = model.fit_predict(image_as_array2d)
```

```python
labels
```

```python
model.cluster_centers_
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
labels
```

```python
rgb_codes[labels]
```

```python
quantised_image = np.reshape(rgb_codes[labels],(h,w,c))
```

```python
quantised_image
```

```python
plt.imshow(quantised_image)
```
